## Filtrando filas con condición

```python
data_c1 = data_raw[data_raw["Day Mins"]>300]
data_c1.shape
```

```python
data_c2= data_raw[data_raw["State"]== "NY"]
data_c2.shape
```

```python
## Concatenación de operaciones booleanas
    ##AND->& OR->|
```

```python
data_c3 = data_raw[(data_raw["Day Mins"]>300) & (data_raw["State"]== "NY")]
data_c3.shape
```

```python
data_c4 = data_raw[(data_raw["Day Mins"]>300) | (data_raw["State"]== "NY")]
data_c4.shape
```
