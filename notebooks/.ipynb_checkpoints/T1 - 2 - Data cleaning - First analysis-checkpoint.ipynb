```python
def createDummies(df,var_name):
    dummy = pd.get_dummies(dataset[var_name], prefix = var_name)
    df = df.drop([var_name],axis=1)
    df = pd.concat([df,dummy],axis=1)
    return df
```

```python
createDummies(dataset,"sex")
```
