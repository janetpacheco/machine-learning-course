# Un dummy data frame con variables categóricas 

## Agregación de datos por categoría

```python
import numpy as np
import pandas as pd
```

```python
gender = [ "Male", "Female"]
socioeconomic_status = ["Poor", "Middle class", "Rich"]
```

```python
n = 500
gender_dataset =[]
socioeconomic_status_dataset = []
for i in range(500):
    gender_dataset.append(np.random.choice(gender))
    socioeconomic_status_dataset.append(np.random.choice(socioeconomic_status))
```

```python
gender_dataset[0:10]
socioeconomic_status_dataset[0:10]
```

```python
#z -> N(0,1)
#N(m,s) ->  m + s *Z
height = 160 + 30*np.random.randn(n)
weight = 60 + 25*np.random.randn(n) 
age = 30 + 12*np.random.randn(n)
income =  18000 + 3500*np.random.randn(n)
```

```python
data = pd.DataFrame(
    {
        "Gender" : gender_dataset,
        "Socioeconomic Status" : socioeconomic_status_dataset,
        "Height" : height,
        "Weight" : weight,
        "Age" : age,
        "Income" : income
    }
)
```

```python
data.
```

## Agrupación de datos

```python
grouped_gender = data.groupby("Gender")
```

```python
grouped_gender.groups
```

```python
for labels,groups in grouped_gender:
    print(labels)
    print(groups)
    
```

```python
grouped_gender.get_group("Female")
```

## Agrupación de más de una categoría

```python
double_group = data.groupby(["Gender", "Socioeconomic Status"])
```

```python
len(double_group)
```

```python
for labels, groups in double_group:
    print(labels)
    print(groups)
```
